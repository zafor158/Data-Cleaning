```python
ages = customer['age']
salaries = customer['salary']

# Check and handle missing values (NaN) in your data
valid_data_mask = ~np.isnan(ages) & ~np.isnan(salaries)
ages = ages[valid_data_mask]
salaries = salaries[valid_data_mask]

# Create a 2D histogram
plt.hist2d(ages, salaries, bins=(20, 20), cmap='YlGnBu')
plt.colorbar(label='Frequency')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs. Salary Histogram')
plt.show()




```

```python
#Group by column names and produce statistical summaries
column_names=['age','workclass','marital-status']
summaries={'salary':'max','education_rank':'min'}
customer=customer.groupby(by=column_names).agg(summaries).reset_index()

#Make sure aggression is done
duplicate=customer.duplicated(subset=column,keep=False)
customer[duplicate]
```

```python
customer.drop_duplicates(inplace=True)
duplicates=customer.duplicated()
customer[customer.duplicated()]
```

```python
age_mean = customer['age'].mean()

customer_imputed = customer.fillna({'age': age_mean})

customer_imputed
```
